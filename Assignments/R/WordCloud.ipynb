```r
library(NLP)
library(tm)
library(RColorBrewer)
library(wordcloud)
```

```r
cv = VCorpus(VectorSource(readLines('./CV.txt')))
```

```r
cv = tm_map(cv, tolower)
cv = tm_map(cv, removePunctuation)
cv = tm_map(cv, removeNumbers)
cv = tm_map(cv, removeWords, stopwords('english'))
cv = tm_map(cv, stripWhitespace)
cv = tm_map(cv, PlainTextDocument)
```

```r
dtm = DocumentTermMatrix(cv)
```

```r
freq = colSums(as.matrix(dtm))
```

```r
freq
```

```r
cvdf = data.frame(var=colnames(as.matrix(dtm)), freq = colSums(as.matrix(dtm)))
```

```r
wordcloud(words =cvdf$var, freq = cvdf$freq,min.freq = 1, scale = c(3,0.5) ,colors = terrain.colors(length(cvdf$var)))
```
