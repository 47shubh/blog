```python
import folium
from xml.dom import minidom

def foliumGPXMap(file_name,line_color = 'black',line_weight = 2.5, line_opacity = 0.5):
    try:
        xmldoc = minidom.parse(file_name)
        itemlist = xmldoc.getElementsByTagName('trkpt')
        lat = [float(s.attributes['lat'].value) for s in itemlist]
        lon = [float(s.attributes['lon'].value) for s in itemlist]
        points = list(map(lambda x,y:(x,y),lat,lon))
        gpxmap = folium.Map(location=[sum(lat)/len(lat), sum(lon)/len(lon)], zoom_start=14)
        folium.PolyLine(points, color = line_color, weight = line_weight, opacity = line_opacity).add_to(gpxmap)
        return gpxmap
    except Exception as e:
        return print('Please check your file and try again...')

foliumGPXMap('./2911108.gpx')
```

```python
import gpxpy
import folium 
 
def foliumGPXMap(file_name,line_color = 'black',line_weight = 2.5, line_opacity = 0.5):
    try:
        with open(file_name, 'r') as file:
            # parsing the gpx file to 
            # GPX(tracks=[GPXTrack(name='...', segments=[GPXTrackSegment(points=[...])])])
            # format and storing it into 'list(tuple())' format in points.  
            gpx = gpxpy.parse(file)
            points = [tuple([point.latitude, point.longitude]) 
              for track in gpx.tracks for segment in track.segments 
              for point in segment.points]
            # Getting the average Latitude & Longitude  for the Map.
            lat = sum(p[0] for p in points)/len(points)
            lon = sum(p[1] for p in points)/len(points)
            gpxmap = folium.Map(location=[lat, lon], zoom_start=14)
            # Adding all points to the map for the given [latitude, longitude].
            folium.PolyLine(points, color = line_color, weight = line_weight, 
                            opacity = line_opacity).add_to(gpxmap)
            return gpxmap
    except Exception as e:
        return print('Please check your file and try again...')

foliumGPXMap('./2911108.gpx')
```
